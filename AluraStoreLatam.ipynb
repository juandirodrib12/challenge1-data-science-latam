### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación


```python
# Calcular el ingreso total de cada tienda
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

print("Ingreso Total por Tienda:")
print(f"Tienda 1: ${ingreso_tienda_1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda_2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda_3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda_4:,.2f}")
```

```python
# Crear un DataFrame con los ingresos totales
import matplotlib.pyplot as plt

datos_facturacion = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
}

df_facturacion = pd.DataFrame(datos_facturacion)
print("\n" + "="*50)
print("Resumen de Facturación by Tienda")
print("="*50)
print(df_facturacion.to_string(index=False))
print("="*50)
print(f"Ingreso Total General: ${df_facturacion['Ingreso Total'].sum():,.2f}")
print("="*50)

# Visualización gráfica
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df_facturacion['Tienda'], df_facturacion['Ingreso Total'], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
ax.set_ylabel('Ingreso Total ($)', fontsize=12)
ax.set_title('Ingreso Total por Tienda', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Agregar valores en las barras
for i, v in enumerate(df_facturacion['Ingreso Total']):
    ax.text(i, v + 100000, f'${v:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y contar ventas para cada tienda
print("="*70)
print("VENTAS POR CATEGORÍA EN CADA TIENDA")
print("="*70)

# Tienda 1
cat_tienda1 = tienda.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 1 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda1.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda1)}")

# Tienda 2
cat_tienda2 = tienda2.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 2 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda2.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda2)}")

# Tienda 3
cat_tienda3 = tienda3.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 3 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda3.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda3)}")

# Tienda 4
cat_tienda4 = tienda4.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 4 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda4.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda4)}")
print("="*70)
```

```python
# Visualización de las categorías más populares por tienda
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Productos Vendidos por Categoría - Comparativa de Tiendas', fontsize=16, fontweight='bold', y=1.00)

# Tienda 1
ax1 = axes[0, 0]
colores = plt.cm.Set3(range(len(cat_tienda1)))
ax1.barh(cat_tienda1['Categoría del Producto'], cat_tienda1['Cantidad de Ventas'], color=colores)
ax1.set_xlabel('Cantidad de Ventas', fontsize=11)
ax1.set_title('TIENDA 1', fontsize=12, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda1['Cantidad de Ventas']):
    ax1.text(v + 1, i, str(v), va='center', fontweight='bold')

# Tienda 2
ax2 = axes[0, 1]
colores = plt.cm.Set2(range(len(cat_tienda2)))
ax2.barh(cat_tienda2['Categoría del Producto'], cat_tienda2['Cantidad de Ventas'], color=colores)
ax2.set_xlabel('Cantidad de Ventas', fontsize=11)
ax2.set_title('TIENDA 2', fontsize=12, fontweight='bold')
ax2.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda2['Cantidad de Ventas']):
    ax2.text(v + 1, i, str(v), va='center', fontweight='bold')

# Tienda 3
ax3 = axes[1, 0]
colores = plt.cm.Pastel1(range(len(cat_tienda3)))
ax3.barh(cat_tienda3['Categoría del Producto'], cat_tienda3['Cantidad de Ventas'], color=colores)
ax3.set_xlabel('Cantidad de Ventas', fontsize=11)
ax3.set_title('TIENDA 3', fontsize=12, fontweight='bold')
ax3.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda3['Cantidad de Ventas']):
    ax3.text(v + 1, i, str(v), va='center', fontweight='bold')

# Tienda 4
ax4 = axes[1, 1]
colores = plt.cm.Pastel2(range(len(cat_tienda4)))
ax4.barh(cat_tienda4['Categoría del Producto'], cat_tienda4['Cantidad de Ventas'], color=colores)
ax4.set_xlabel('Cantidad de Ventas', fontsize=11)
ax4.set_title('TIENDA 4', fontsize=12, fontweight='bold')
ax4.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda4['Cantidad de Ventas']):
    ax4.text(v + 1, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Resumen de categorías más populares
print("\n" + "="*70)
print("CATEGORÍA MÁS POPULAR EN CADA TIENDA")
print("="*70)
print(f"Tienda 1: {cat_tienda1.iloc[0]['Categoría del Producto']} ({cat_tienda1.iloc[0]['Cantidad de Ventas']} ventas)")
print(f"Tienda 2: {cat_tienda2.iloc[0]['Categoría del Producto']} ({cat_tienda2.iloc[0]['Cantidad de Ventas']} ventas)")
print(f"Tienda 3: {cat_tienda3.iloc[0]['Categoría del Producto']} ({cat_tienda3.iloc[0]['Cantidad de Ventas']} ventas)")
print(f"Tienda 4: {cat_tienda4.iloc[0]['Categoría del Producto']} ({cat_tienda4.iloc[0]['Cantidad de Ventas']} ventas)")
print("="*70)
```
