### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación


```python
# Calcular el ingreso total de cada tienda
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

print("Ingreso Total por Tienda:")
print(f"Tienda 1: ${ingreso_tienda_1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda_2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda_3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda_4:,.2f}")
```

```python
# Crear un DataFrame con los ingresos totales
import matplotlib.pyplot as plt

datos_facturacion = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
}

df_facturacion = pd.DataFrame(datos_facturacion)
print("\n" + "="*50)
print("Resumen de Facturación by Tienda")
print("="*50)
print(df_facturacion.to_string(index=False))
print("="*50)
print(f"Ingreso Total General: ${df_facturacion['Ingreso Total'].sum():,.2f}")
print("="*50)

# Visualización gráfica
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df_facturacion['Tienda'], df_facturacion['Ingreso Total'], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
ax.set_ylabel('Ingreso Total ($)', fontsize=12)
ax.set_title('Ingreso Total por Tienda', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Agregar valores en las barras
for i, v in enumerate(df_facturacion['Ingreso Total']):
    ax.text(i, v + 100000, f'${v:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y contar ventas para cada tienda
print("="*70)
print("VENTAS POR CATEGORÍA EN CADA TIENDA")
print("="*70)

# Tienda 1
cat_tienda1 = tienda.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 1 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda1.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda1)}")

# Tienda 2
cat_tienda2 = tienda2.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 2 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda2.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda2)}")

# Tienda 3
cat_tienda3 = tienda3.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 3 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda3.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda3)}")

# Tienda 4
cat_tienda4 = tienda4.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 4 - Cantidad de Productos Vendidos por Categoría:")
print(cat_tienda4.to_string(index=False))
print(f"Total de categorías: {len(cat_tienda4)}")
print("="*70)
```

```python
# Visualización de las categorías más populares por tienda
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Productos Vendidos por Categoría - Comparativa de Tiendas', fontsize=16, fontweight='bold', y=1.00)

# Tienda 1
ax1 = axes[0, 0]
colores = plt.cm.Set3(range(len(cat_tienda1)))
ax1.barh(cat_tienda1['Categoría del Producto'], cat_tienda1['Cantidad de Ventas'], color=colores)
ax1.set_xlabel('Cantidad de Ventas', fontsize=11)
ax1.set_title('TIENDA 1', fontsize=12, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda1['Cantidad de Ventas']):
    ax1.text(v + 1, i, str(v), va='center', fontweight='bold')

# Tienda 2
ax2 = axes[0, 1]
colores = plt.cm.Set2(range(len(cat_tienda2)))
ax2.barh(cat_tienda2['Categoría del Producto'], cat_tienda2['Cantidad de Ventas'], color=colores)
ax2.set_xlabel('Cantidad de Ventas', fontsize=11)
ax2.set_title('TIENDA 2', fontsize=12, fontweight='bold')
ax2.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda2['Cantidad de Ventas']):
    ax2.text(v + 1, i, str(v), va='center', fontweight='bold')

# Tienda 3
ax3 = axes[1, 0]
colores = plt.cm.Pastel1(range(len(cat_tienda3)))
ax3.barh(cat_tienda3['Categoría del Producto'], cat_tienda3['Cantidad de Ventas'], color=colores)
ax3.set_xlabel('Cantidad de Ventas', fontsize=11)
ax3.set_title('TIENDA 3', fontsize=12, fontweight='bold')
ax3.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda3['Cantidad de Ventas']):
    ax3.text(v + 1, i, str(v), va='center', fontweight='bold')

# Tienda 4
ax4 = axes[1, 1]
colores = plt.cm.Pastel2(range(len(cat_tienda4)))
ax4.barh(cat_tienda4['Categoría del Producto'], cat_tienda4['Cantidad de Ventas'], color=colores)
ax4.set_xlabel('Cantidad de Ventas', fontsize=11)
ax4.set_title('TIENDA 4', fontsize=12, fontweight='bold')
ax4.grid(axis='x', alpha=0.3)
for i, v in enumerate(cat_tienda4['Cantidad de Ventas']):
    ax4.text(v + 1, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
plt.show()

# Resumen de categorías más populares
print("\n" + "="*70)
print("CATEGORÍA MÁS POPULAR EN CADA TIENDA")
print("="*70)
print(f"Tienda 1: {cat_tienda1.iloc[0]['Categoría del Producto']} ({cat_tienda1.iloc[0]['Cantidad de Ventas']} ventas)")
print(f"Tienda 2: {cat_tienda2.iloc[0]['Categoría del Producto']} ({cat_tienda2.iloc[0]['Cantidad de Ventas']} ventas)")
print(f"Tienda 3: {cat_tienda3.iloc[0]['Categoría del Producto']} ({cat_tienda3.iloc[0]['Cantidad de Ventas']} ventas)")
print(f"Tienda 4: {cat_tienda4.iloc[0]['Categoría del Producto']} ({cat_tienda4.iloc[0]['Cantidad de Ventas']} ventas)")
print("="*70)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio para cada tienda
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print("="*50)
print("CALIFICACIÓN PROMEDIO POR TIENDA")
print("="*50)
print(f"Tienda 1: {calificacion_tienda1:.2f}")
print(f"Tienda 2: {calificacion_tienda2:.2f}")
print(f"Tienda 3: {calificacion_tienda3:.2f}")
print(f"Tienda 4: {calificacion_tienda4:.2f}")
print("="*50)
print(f"Promedio General: {(calificacion_tienda1 + calificacion_tienda2 + calificacion_tienda3 + calificacion_tienda4) / 4:.2f}")
print("="*50)
```

```python
# Crear un DataFrame con las calificaciones promedio y visualizar
calificaciones_prom = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
}

df_calif = pd.DataFrame(calificaciones_prom)

# Ordenar por calificación descendente
df_calif = df_calif.sort_values('Calificación Promedio', ascending=False).reset_index(drop=True)

print("\n" + "="*50)
print("Resumen de Calificación Promedio por Tienda")
print("="*50)
print(df_calif.to_string(index=False))
print("="*50)

# Visualización - Gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Crear barras con colores degradados según calificación
colores_barras = ['#2ca02c' if x >= 4.5 else '#ff7f0e' if x >= 4.0 else '#d62728' for x in df_calif['Calificación Promedio']]

barras = ax.bar(range(len(df_calif)), df_calif['Calificación Promedio'], color=colores_barras, edgecolor='black', linewidth=1.5)

# Personalización del gráfico
ax.set_ylabel('Calificación Promedio', fontsize=12, fontweight='bold')
ax.set_title('Calificación Promedio por Tienda', fontsize=14, fontweight='bold')
ax.set_xticks(range(len(df_calif)))
ax.set_xticklabels(df_calif['Tienda'], fontsize=11)
ax.set_ylim(0, 5)
ax.axhline(y=4.0, color='orange', linestyle='--', linewidth=2, label='Meta: 4.0 ⭐', alpha=0.7)
ax.axhline(y=4.5, color='green', linestyle='--', linewidth=2, label='Excelente: 4.5 ⭐', alpha=0.7)
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.legend(loc='lower right', fontsize=10)

# Agregar valores en las barras
for i, (barra, valor) in enumerate(zip(barras, df_calif['Calificación Promedio'])):
    altura = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2., altura + 0.1,
            f'{valor:.2f}',
            ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Análisis de productos más y menos vendidos por tienda
print("="*80)
print("PRODUCTOS MÁS Y MENOS VENDIDOS POR TIENDA")
print("="*80)

# Tienda 1
prod_tienda1 = tienda.groupby('Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\nTIENDA 1 - Top 5 Productos Más Vendidos:")
print(prod_tienda1.head(5).to_string(index=False))
print("\nTIENDA 1 - Top 5 Productos Menos Vendidos:")
print(prod_tienda1.tail(5).to_string(index=False))

# Tienda 2
prod_tienda2 = tienda2.groupby('Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\n" + "-"*80)
print("\nTIENDA 2 - Top 5 Productos Más Vendidos:")
print(prod_tienda2.head(5).to_string(index=False))
print("\nTIENDA 2 - Top 5 Productos Menos Vendidos:")
print(prod_tienda2.tail(5).to_string(index=False))

# Tienda 3
prod_tienda3 = tienda3.groupby('Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\n" + "-"*80)
print("\nTIENDA 3 - Top 5 Productos Más Vendidos:")
print(prod_tienda3.head(5).to_string(index=False))
print("\nTIENDA 3 - Top 5 Productos Menos Vendidos:")
print(prod_tienda3.tail(5).to_string(index=False))

# Tienda 4
prod_tienda4 = tienda4.groupby('Producto').size().reset_index(name='Cantidad de Ventas').sort_values('Cantidad de Ventas', ascending=False)
print("\n" + "-"*80)
print("\nTIENDA 4 - Top 5 Productos Más Vendidos:")
print(prod_tienda4.head(5).to_string(index=False))
print("\nTIENDA 4 - Top 5 Productos Menos Vendidos:")
print(prod_tienda4.tail(5).to_string(index=False))
print("\n" + "="*80)
```

```python
# Visualización de productos más y menos vendidos
fig, axes = plt.subplots(4, 2, figsize=(18, 16))
fig.suptitle('Productos Más y Menos Vendidos por Tienda', fontsize=16, fontweight='bold', y=0.995)

tiendas_data = [
    ('TIENDA 1', prod_tienda1, 0),
    ('TIENDA 2', prod_tienda2, 1),
    ('TIENDA 3', prod_tienda3, 2),
    ('TIENDA 4', prod_tienda4, 3)
]

for tienda_name, prod_data, idx in tiendas_data:
    # Top 5 más vendidos
    top_5 = prod_data.head(5)
    ax_top = axes[idx, 0]
    colores_top = ['#2ca02c' if i == 0 else '#1f77b4' for i in range(len(top_5))]
    ax_top.barh(range(len(top_5)), top_5['Cantidad de Ventas'].values, color=colores_top, edgecolor='black', linewidth=1.2)
    ax_top.set_yticks(range(len(top_5)))
    ax_top.set_yticklabels(top_5['Producto'].values, fontsize=9)
    ax_top.set_xlabel('Cantidad de Ventas', fontsize=10, fontweight='bold')
    ax_top.set_title(f'{tienda_name} - Top 5 Más Vendidos', fontsize=11, fontweight='bold', color='darkgreen')
    ax_top.grid(axis='x', alpha=0.3)
    ax_top.invert_yaxis()
    
    # Agregar valores
    for i, v in enumerate(top_5['Cantidad de Ventas'].values):
        ax_top.text(v + 0.5, i, str(int(v)), va='center', fontweight='bold', fontsize=9)
    
    # Top 5 menos vendidos
    bottom_5 = prod_data.tail(5).sort_values('Cantidad de Ventas', ascending=True)
    ax_bottom = axes[idx, 1]
    colores_bottom = ['#d62728' if i == len(bottom_5)-1 else '#ff7f0e' for i in range(len(bottom_5))]
    ax_bottom.barh(range(len(bottom_5)), bottom_5['Cantidad de Ventas'].values, color=colores_bottom, edgecolor='black', linewidth=1.2)
    ax_bottom.set_yticks(range(len(bottom_5)))
    ax_bottom.set_yticklabels(bottom_5['Producto'].values, fontsize=9)
    ax_bottom.set_xlabel('Cantidad de Ventas', fontsize=10, fontweight='bold')
    ax_bottom.set_title(f'{tienda_name} - Top 5 Menos Vendidos', fontsize=11, fontweight='bold', color='darkred')
    ax_bottom.grid(axis='x', alpha=0.3)
    
    # Agregar valores
    for i, v in enumerate(bottom_5['Cantidad de Ventas'].values):
        ax_bottom.text(v + 0.1, i, str(int(v)), va='center', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()

# Resumen general
print("\n" + "="*80)
print("RESUMEN - PRODUCTOS ESTRELLA POR TIENDA")
print("="*80)
for tienda_name, prod_data, _ in tiendas_data:
    producto_top = prod_data.iloc[0]['Producto']
    ventas_top = prod_data.iloc[0]['Cantidad de Ventas']
    producto_bottom = prod_data.iloc[-1]['Producto']
    ventas_bottom = prod_data.iloc[-1]['Cantidad de Ventas']
    
    print(f"\n{tienda_name}:")
    print(f"  ー Producto MÁS vendido: {producto_top} ({ventas_top} unidades)")
    print(f"  ー Producto MENOS vendido: {producto_bottom} ({ventas_bottom} unidades)")
print("="*80)
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
costo_envio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_tienda4 = tienda4['Costo de envío'].mean()

print("="*60)
print("COSTO DE ENVÍO PROMEDIO POR TIENDA")
print("="*60)
print(f"Tienda 1: ${costo_envio_tienda1:,.2f}")
print(f"Tienda 2: ${costo_envio_tienda2:,.2f}")
print(f"Tienda 3: ${costo_envio_tienda3:,.2f}")
print(f"Tienda 4: ${costo_envio_tienda4:,.2f}")
print("="*60)
print(f"Promedio General: ${(costo_envio_tienda1 + costo_envio_tienda2 + costo_envio_tienda3 + costo_envio_tienda4) / 4:,.2f}")
print("="*60)

# Calcular ingresos por envío y ahorros potenciales
total_envios_t1 = tienda['Costo de envío'].sum()
total_envios_t2 = tienda2['Costo de envío'].sum()
total_envios_t3 = tienda3['Costo de envío'].sum()
total_envios_t4 = tienda4['Costo de envío'].sum()

print("\nINGRESOS POR ENVÍO (Lo que pagan los clientes):")
print("="*60)
print(f"Tienda 1: ${total_envios_t1:,.2f} ({len(tienda)} transacciones)")
print(f"Tienda 2: ${total_envios_t2:,.2f} ({len(tienda2)} transacciones)")
print(f"Tienda 3: ${total_envios_t3:,.2f} ({len(tienda3)} transacciones)")
print(f"Tienda 4: ${total_envios_t4:,.2f} ({len(tienda4)} transacciones)")
print("="*60)
print(f"Total General: ${total_envios_t1 + total_envios_t2 + total_envios_t3 + total_envios_t4:,.2f}")
print("="*60)
```

```python
# Crear DataFrames para visualización
datos_envio = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo Promedio': [costo_envio_tienda1, costo_envio_tienda2, costo_envio_tienda3, costo_envio_tienda4],
    'Total Ingresos Envío': [total_envios_t1, total_envios_t2, total_envios_t3, total_envios_t4],
    'Cantidad Transacciones': [len(tienda), len(tienda2), len(tienda3), len(tienda4)]
}

df_envio = pd.DataFrame(datos_envio)

print("\n" + "="*80)
print("Resumen de Costos de Envío por Tienda")
print("="*80)
print(df_envio.to_string(index=False))
print("="*80)

# Visualización - Gráfico comparativo
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Análisis de Costos de Envío por Tienda', fontsize=14, fontweight='bold')

# Gráfico 1: Costo promedio de envío
tiendas = df_envio['Tienda']
costos_promedio = df_envio['Costo Promedio']
colores1 = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

barras1 = ax1.bar(tiendas, costos_promedio, color=colores1, edgecolor='black', linewidth=1.5, alpha=0.8)
ax1.set_ylabel('Costo Promedio ($)', fontsize=11, fontweight='bold')
ax1.set_title('Costo de Envío Promedio por Tienda', fontsize=12, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)

# Valores en las barras
for barra, valor in zip(barras1, costos_promedio):
    altura = barra.get_height()
    ax1.text(barra.get_x() + barra.get_width()/2., altura + 200,
            f'${valor:,.0f}',
            ha='center', va='bottom', fontweight='bold', fontsize=10)

# Línea de promedio general
promedio_general = costos_promedio.mean()
ax1.axhline(y=promedio_general, color='red', linestyle='--', linewidth=2, label=f'Promedio: ${promedio_general:,.0f}', alpha=0.7)
ax1.legend(fontsize=10)

# Gráfico 2: Ingresos totales por envío
ingresos_envio = df_envio['Total Ingresos Envío']
barras2 = ax2.bar(tiendas, ingresos_envio, color=colores1, edgecolor='black', linewidth=1.5, alpha=0.8)
ax2.set_ylabel('Ingresos Totales ($)', fontsize=11, fontweight='bold')
ax2.set_title('Ingresos Totales por Envíos Pagados por Clientes', fontsize=12, fontweight='bold')
ax2.grid(axis='y', alpha=0.3)

# Valores en las barras
for barra, valor in zip(barras2, ingresos_envio):
    altura = barra.get_height()
    ax2.text(barra.get_x() + barra.get_width()/2., altura + 1000000,
            f'${valor/1000000:.1f}M',
            ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()

# Análisis adicional
print("\n" + "="*80)
print("ANÁLISIS Y RECOMENDACIONES")
print("="*80)
tienda_mas_alto = df_envio.loc[df_envio['Costo Promedio'].idxmax()]
tienda_mas_bajo = df_envio.loc[df_envio['Costo Promedio'].idxmin()]

print(f"\nCosto de envío MÁS ALTO: {tienda_mas_alto['Tienda']} (${tienda_mas_alto['Costo Promedio']:,.2f})")
print(f"Costo de envío MÁS BAJO: {tienda_mas_bajo['Tienda']} (${tienda_mas_bajo['Costo Promedio']:,.2f})")
print(f"Diferencia: ${tienda_mas_alto['Costo Promedio'] - tienda_mas_bajo['Costo Promedio']:,.2f}")

print(f"\nTienda con mayores ingresos por envío: {df_envio.loc[df_envio['Total Ingresos Envío'].idxmax()]['Tienda']}")
print(f"  Monto: ${df_envio['Total Ingresos Envío'].max():,.2f}")
print("="*80)
```
